```python
import requests
import json5
import time
import pandas as pd
```

```python
jobs = []
```

```python
for pg in range(11596//50):
    r = requests.get('https://www.indeed.com/jobs?q&l=Rancho%20Cordova%2C%20CA&radius=15&fromage=14&start={}&limit=50'.format(pg*50))
    if r.status_code != 200:
        print('Error',pg)
        time.sleep(5)
        continue
    for i in range(50):
        start = r.text.find('jobmap[{}]'.format(i))
        if start==-1:
            continue
        end = r.text.find('jobmap[{}]'.format(i+1))
        s = r.text.find('{',start)
        e = r.text.find('}',s)+1
        job = r.text[s:e]
        p = job.find(': ')
        while p != -1:
            job = job.replace(': ',':')
            p = job.find(': ')
        job = json5.loads(job)
        jobs.append(job)
    time.sleep(5)
```

```python
jobs
```

```python
df = pd.DataFrame(jobs)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.to_csv('jobs.csv',index=False)
```
