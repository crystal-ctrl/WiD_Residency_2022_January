```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Hide Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../Data/linkedin_job_posts_cleanish.csv')
df.head()
```

```python
df.drop(columns=['Job Link','Unnamed: 0'], inplace=True)
df.head()
```

```python
jobs_by_seniority = df.groupby(['Seniority level'])['Title'].count().sort_values(ascending=False)
jobs_by_seniority
```

```python
jobs_by_seniority.sort_values().plot(kind='bar')
plt.title("Seniority level with most job openings")
```
